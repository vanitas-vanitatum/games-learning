```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.figure(figsize=(16, 12))

aa = pd.read_csv('deepqlearning_analysis.csv')[['MeanRewardO']]
aa.plot(figsize=(16, 12))
```

```python
sliding_window_results = []
for i in range(0, len(aa) - 2):
    sliding_window_results.append({
        'Mean reward': np.mean(aa.iloc[i:i + 2]['MeanRewardO'])
    })
pd.DataFrame(sliding_window_results).plot()
```
